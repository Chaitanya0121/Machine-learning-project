```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv("C:\\Users\\Chaitanya\\OneDrive\\Desktop\\AI Lab\\Lab\\Project\\Dataset\\carDekho Dataset\\Car details v3 24July.csv")     
data
```

```python
x = data[['year','km_driven','mileage (kmpl)','engine (cc)','seats']]
y = data[['selling_price']]
```

```python
sns.pairplot(data, x_vars=['year','km_driven','mileage (kmpl)','engine (cc)','seats'], y_vars='selling_price')
```

```python
min_max_scaler = MinMaxScaler().fit(x)
x_norm = min_max_scaler.transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_norm,y,test_size=0.3,random_state=800)
```

```python
LR = LinearRegression()
LR.fit(x_test,y_test)
List = []
List = LR.coef_.tolist()
C = LR.intercept_
```

```python
y_pred = LR.predict(x_train)
score = LR.score(x_test,y_test)
print("Accuracy Score: ",score)
```
